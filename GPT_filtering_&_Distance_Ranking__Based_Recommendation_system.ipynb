##**Distance-Ranking based Hotel-Restaurant Recommendation System**



```python
import pandas as pd
from itertools import product
```

```python

# Read the hotel and restaurant datasets
hotel_df = pd.read_csv("/content/hotels_gpt3_final.csv")
restaurant_df = pd.read_csv("/content/restaurants_gpt3_final.csv")
```

```python
hotel_df
```

```python
restaurant_df
```

```python
# Remove units from Distance
hotel_df['Distance'] = hotel_df['Distance'].str.replace(' km', '')

# Remove units from Price per day
hotel_df['Price per day'] = hotel_df['Price per day'].str.replace(' USD', '')

# Remove units from Ratings
hotel_df['Ratings'] = hotel_df['Ratings'].str.replace(' stars', '')
```

```python
# Convert miles to kilometers
restaurant_df.loc[restaurant_df['Unit'] == 'miles', 'Distance'] = restaurant_df.loc[restaurant_df['Unit'] == 'miles', 'Distance'] * 1.60934

# Update the unit to km
restaurant_df.loc[restaurant_df['Unit'] == 'miles', 'Unit'] = 'km'

# Filter for Chinese and Italian cuisines
restaurant_df = restaurant_df[restaurant_df['Cuisine'].isin(['Chinese', 'Italian'])]

# Print the filtered DataFrame
print(restaurant_df)
```

```python
# Rename "Name" column to "Hotel Name"
hotel_df=hotel_df.rename(columns={'Name': 'Hotel Name'})

# Rename "Address" column to "Restaurant"
restaurant_df=restaurant_df.rename(columns={'Name': 'Restaurant Name'})
```

```python
hotel_df.head()
```

```python
restaurant_df.head()
```

```python

# Calculate hotel and restaurant rankings
hotel_df['Hotel_Rank'] = hotel_df['Ratings'].rank(ascending=False)
restaurant_df['Restaurant_Rank'] = restaurant_df['Rating'].rank(ascending=False)
```

```python
# Calculate distances between all possible hotel-restaurant pairs
distances = []
for hotel, restaurant in product(hotel_df.iterrows(), restaurant_df.iterrows()):
    hotel_index, hotel_data = hotel
    restaurant_index, restaurant_data = restaurant
    distance = abs(hotel_data['Distance'] - restaurant_data['Distance'])
    distances.append((hotel_index, restaurant_index, distance))
```

```python
# Creating dataFrame with the distances
distance_df = pd.DataFrame(distances, columns=['Hotel_Index', 'Restaurant_Index', 'Distance'])
```

```python
# Merge the distance DataFrame with the hotel and restaurant DataFrames
merged_df = pd.merge(distance_df, hotel_df, left_on='Hotel_Index', right_index=True)
merged_df = pd.merge(merged_df, restaurant_df, left_on='Restaurant_Index', right_index=True)
```

```python
# Calculate the combined rank using weighted average
weight_distance = 0.4
weight_hotel_rank = 0.3
weight_restaurant_rank = 0.3

merged_df['Combined_Rank'] = (weight_distance * merged_df['Distance']) + \
                             (weight_hotel_rank * merged_df['Hotel_Rank']) + \
                             (weight_restaurant_rank * merged_df['Restaurant_Rank'])
```

```python
# Sort the combined rank in ascending order to get the top recommendation
best_pair = merged_df.sort_values('Combined_Rank').iloc[0]
```

```python

# Extract the relevant information
best_hotel_name = best_pair['Hotel Name']
best_restaurant_name = best_pair['Restaurant Name']
combined_rank = best_pair['Combined_Rank']

# Display the best hotel and restaurant recommendation
print("Best Hotel and Restaurant Recommendation:")
print("Hotel Name:", best_hotel_name)
print("Restaurant Name:", best_restaurant_name)

```

##**This recommendation system takes into account the rankings (ratings) of hotels and restaurants as well as the distances from the stadium to suggest the best hotel-restaurant pair that combines both favorable rankings and proximity to the stadium.**
